# Login to your Instagram Handle

```python
def LogIn(username, password):
    try :
        user_name = wait.until(EC.presence_of_element_located((By.NAME,'username')))
        user_name.send_keys(username)
        #locating password box and sending password
        pwd = driver.find_element_by_name('password')
        pwd.send_keys(password)
        #locating login button 
        button = wait.until(EC.presence_of_element_located((By.XPATH,'//*[@id="loginForm"]/div[1]/div[3]/button/div')))
        button.submit()
        #Save Your Login Info? : Not Now
        pop = wait.until(EC.presence_of_element_located((By.XPATH,'//*[@id="react-root"]/section/main/div/div/div/div/button')))
        pop.click()
        #TurnONnotifications
        notifications = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_a9_0')))
        notifications.click()
    except TimeoutException :
        print ("Something went wrong! Try Again")
```

# Follow/Unfollow given handle - 

```python
def follow():
    try :
        #locating follow button
        btn = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_aad6')))
        #checking for text
        if btn.text == 'Follow' : 
            btn.click()
            time.sleep(3)
        
        else : 
            print('Already Following')
            
        
        
    except TimeoutException :
        print("Something Went Wrong!")
 
```

```python
def unfollow():
    try :
        #locating follow button
        btn = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_aad6')))
        #checking for text
        if btn.text !='Follow' : 
            btn.click()
            time.sleep(2)
            #locating popup window (when you click on follow button)
            pop_up = wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_a9-_')))
            pop_up.click()
            time.sleep(3)
            
        else : 
            print('Already Unfollowed')
        

    except TimeoutException :
        print("Something Went Wrong!")
```

# Like/Unlike posts

```python
def Like_Post():
    try :
        #scrolling for locating post
        driver.execute_script('window.scrollTo(0, 6000);')
        time.sleep(3)
        driver.execute_script('window.scrollTo(0, -6000);')
        time.sleep(3)
        #locating post
        posts = driver.find_elements_by_class_name('_aa8k')
        
        for i in range(30):
            posts[i].click()
            time.sleep(2)
            #locating like/unke button
            like = driver.find_element(By.CLASS_NAME,'_abl_') 
            st = BeautifulSoup(like.get_attribute('innerHTML'),"html.parser").svg['aria-label']
        
            if st == 'Like' :
                like.click()
                time.sleep(2)
            else :
                print('You have already LIKED Post Number :', i+1)
                time.sleep(2)
            #locating cross button for closing post
            driver.back()
            time.sleep(2)
            
    except TimeoutException :
        print("Something Went Wrong!")
```

```python
def Unlike_Post():
    try :
        #scrolling for locating post
        driver.execute_script('window.scrollTo(0, 6000);')
        time.sleep(3)
        driver.execute_script('window.scrollTo(0, -6000);')
        time.sleep(3)
        #locating post
        posts = driver.find_elements_by_class_name('_aa8k')
     
        for i in range(30):
            posts[i].click()
            time.sleep(2)
            #locating like/unke button
            like = wait.until(EC.presence_of_element_located((By.XPATH,'//section[contains(@class,"_aamu")]//div[contains(@class,"_abm0")]/span')))
            st = BeautifulSoup(like.get_attribute('innerHTML'),"html.parser").svg['aria-label']
            
            if st == 'Unlike' :
                like.click()
                time.sleep(2)
            else :
                print('You have already UNLIKED Post Number', i+1)
                time.sleep(2)
            #locating cross button for closing post
            driver.back()
            time.sleep(2)
            
    except TimeoutException :
        print("Something Went Wrong!")
```

```python
def Extract_Followers():
    try :
        follower_button = driver.find_element(by = By.CLASS_NAME, value = 'oajrlxb2')
        follower_button.click()
        
        time.sleep(3)
        
        frame = driver.find_element_by_class_name('_aano')
                #scrolling untill first 500 user is located
        for i in range(60):
            time.sleep(1.5)
            driver.execute_script("arguments[0].scrollTop=arguments[0].scrollHeight;",frame)

        names = []
        #extracting userdata
        followers = driver.find_elements_by_class_name('_aaei')
        #extracting username
        for i in followers[:500] :
            names.append(i.text.split('\n')[0])  
        return names
    except TimeoutException :
        print("Something Went Wrong!")
```

```python
def Following():
    try :
        # locating following button and click on it
        following_button = driver.find_elements(by = By.CLASS_NAME, value = 'oajrlxb2')
        following_button[2].click()
        
        time.sleep(3)
        
        frame = driver.find_element_by_class_name('_aano')
        #scrolling untill all user is located
        for i in range(30):
            time.sleep(1)
            driver.execute_script("arguments[0].scrollTop=arguments[0].scrollHeight;",frame)


        names = []
        #extracting userdata
        following = driver.find_elements_by_class_name('_aaei')
        #extracting username
        for i in following :
                names.append(i.text.split('\n')[0])
            
        return names
    except TimeoutException :
        print("Something Went Wrong!")
```

# Check the story of ‘coding.ninjas’. Consider the following Scenarios and print error messages accordingly 

```python
def Check_Story():
    try:
        #locating story or profile pic
        story = wait.until(EC.presence_of_element_located((By.CLASS_NAME,"_aarf")))
    
        #check the Profile photo size to find out story is available or not
        e = driver.find_element_by_xpath("//div[contains(@class,'_aarf')]/canvas")
        size = e.size
        height = size['height']
        
        if int(height) == 166:
            print("Already seen the story")
        else:
            print("Viewing the story")
            driver.execute_script("arguments[0].click();",story)
    except:
        print("No story is available to view")
```
